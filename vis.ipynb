```python
def vis(target_date, window_range, target_region, event_time = None, event_name = None, news_keyword = None):
    target_time = datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S')
    target_day = (target_time - start_time).days
    idx = code2idx[region2code[target_region]]
    
    #(datetime.strptime(event_time, '%Y-%m-%d %H:%M:%S') - start_time
    
    # 示例数据
    data = [i for day_flow in flows[target_day-window_range:target_day+window_range, :,idx] for i in day_flow.tolist()]

    # 创建一个新的图形
    plt.figure(figsize=(32, 8))

    # 绘制折线图
    plt.plot(data, lw=4, label="Human Flow")

    # 添加标题和标签
    #plt.title('Noto Earthquake')
    plt.xlabel('Time', fontsize=25)
    #plt.ylabel('Human Flow', fontsize=25)

    #plt.axvline(x=6 * 96 + 16 * 4, color='r', linestyle='--')
    for i in range(2 * window_range + 1):
        plt.axvline(x=i * 96, color='green', linestyle='--', lw=4)
    
    event_idx = window_range * 96 + (datetime.strptime(event_time, '%Y-%m-%d %H:%M:%S') - datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S')).total_seconds() // 900
    plt.axvline(x=event_idx, color='red', linestyle='dotted', lw=5, label = event_name, alpha = 0.5)
    
    x_positions = [i * 96 for i in range(2 * window_range + 1)]
    x_labels = [(datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S') + timedelta(days=i)).strftime('%Y-%m-%d') + ' ' + (datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S') + timedelta(days=i)).strftime('%a') for i in range(-window_range, window_range + 1)]
    plt.xticks(ticks=x_positions, labels=x_labels, rotation=45, fontsize=20)
    plt.yticks(fontsize=20)
    
    # 显示图形
    

    x_labels = [(datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S') + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(-window_range, window_range)]
    filtered_news = news[
        (news['DateAndTime'] >= pd.to_datetime(x_labels[0])) & 
        (news['DateAndTime'] <= pd.to_datetime(x_labels[-1])) &
        (news['codes'].str.contains(region2code[target_region]))
    ]
    filtered_news = filtered_news.sort_values(by='DateAndTime') 
    filtered_news = filtered_news[filtered_news['Content'].str.contains(news_keyword)]
    if len(filtered_news) > 0:
        news_time = filtered_news.iloc[0]['DateAndTime']
        news_idx = window_range * 96 + (news_time - datetime.strptime(target_date, '%Y-%m-%d %H:%M:%S')).total_seconds() // 900
        plt.axvline(x=news_idx, color='orange', linestyle='--', lw=5, label = 'News Released Time', alpha = 0.5)
    
    plt.legend(fontsize=20)
    plt.show()
    return filtered_news
```

```python
#隅田川花火大会
target_date = '2024-01-01 00:00:00'
event_time = '2024-01-01 16:06:00'
event_name = 'Noto Earthquake'
news_keyword = '震度６'
window_range = 3
target_region = '石川県金沢市'
returned_news = vis(target_date, window_range, target_region, event_time, event_name, news_keyword)
```
