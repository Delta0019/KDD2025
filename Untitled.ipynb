```python
import time
import numpy as np
import torch
from sklearn.metrics.pairwise import cosine_similarity

# 初始化数据
data = torch.rand((32, 490, 16))
base = torch.rand((1, 16))

# 记录开始时间
start_time = time.time()

# 数据变形
data = data.reshape(32 * 490, 16)
batch_size = 32

# 将数据从 GPU 移到 CPU 并转换为 numpy 数组
data = data.cpu().detach().numpy()
base = base.cpu().detach().numpy()

# 计算余弦相似度
similarities = cosine_similarity(data, base)
similarities = similarities.reshape(batch_size, -1)

# 分组和标签分配
batch_labels = []
for i in range(batch_size):
    similarity = similarities[i]
    bins = np.arange(0, 1.1, 0.1)
    labels = np.digitize(similarity, bins, right=True)
    unique_labels = np.unique(labels)
    label_mapping = {label: idx for idx, label in enumerate(unique_labels)}
    new_labels = np.array([label_mapping[label] for label in labels])
    batch_labels.append(new_labels)

# 记录结束时间
end_time = time.time()

# 计算并打印耗时
elapsed_time = end_time - start_time
print(f"Total time elapsed: {elapsed_time:.4f} seconds")
```
