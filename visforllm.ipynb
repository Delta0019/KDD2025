```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import interp2d
from collections import Counter

def interpolation(x, z):
    interp_func = interp2d(x, np.arange(len(z)), z, kind='linear')

    # 生成插值后的x值
    x_interp = np.arange(x.min(), x.max() + 0.001, 0.001)
    
    # 计算插值后的z值
    z_interp = interp_func(x_interp, np.arange(len(z)))
    
    # 确保所有z值非负
    z_interp = np.maximum(z_interp, 0)
    z_interp = np.minimum(z_interp, 1)
    
    return x_interp, z_interp
```

```python
# 创建3D图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 创建带色彩的ribbon图

colors = plt.cm.Paired(np.linspace(0, 1, scores.shape[1] - 1))


for i in range(scores.shape[1] - 1):
    frequency = Counter(scores[:, i])
    x = np.abs(sorted(frequency.keys()))
    
    z = np.array([frequency[num]/scores.shape[0] for num in x])
    #x, z = interpolation(x, z)
    y = 2 * np.array([[i]] * len(x))
    ax.plot(100 * x, y, z, color=colors[i])


# 设置标签
ax.set_xlabel('Scores')
#ax.set_ylabel('Types')
ax.set_zlabel('Frequency')

#plt.title('三维带色彩的ribbon图')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = scores[:, :-1]

# 定义区间边界
bins = np.arange(0, 1.1, 0.1)

# 创建2D图
fig, ax = plt.subplots(figsize=(12, 8))

# 颜色列表
colors = plt.cm.jet(np.linspace(0, 1, data.shape[1]))

# 计算每个维度的频率并绘制图形
for i in range(data.shape[1]):
    counts, _ = np.histogram(np.abs(data[:, i]), bins=bins)
    counts = counts / len(data)  # 计算每个区间的占比
    print(i, counts)
    x = (bins[:-1] + bins[1:]) / 2  # 计算每个区间的中心点作为x轴

    # 绘制折线图
    ax.bar(x, counts)

# 设置标签
# ax.set_xlabel('Value Range')
# ax.set_ylabel('Proportion')
# ax.set_title('Proportion of Different Intervals for Each Dimension')
ax.legend()

plt.show()
```

```python
freq = {}
for i in range(scores.shape[1]):
    feat = np.abs(scores[:, i])
    freq[i] = {}
    for j in range(0, 10):
        if j != 9:
            count = np.sum((feat >= j * 0.1) & (feat < (j+1) * 0.1))
        else:
            count = np.sum((feat >= j * 0.1) & (feat <= (j+1) * 0.1))
        freq[i][round(j*0.1, 1)] = count / scores.shape[0]
```
